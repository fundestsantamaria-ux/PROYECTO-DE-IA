```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
from IPython.display import HTML, display
import re

CLIENT_ID = "904927fbd688484dbbd8f211a71a7682"
CLIENT_SECRET = "77503841ff704b45bbe7cf56e71a05c0"

sp = spotipy.Spotify(
    auth_manager=SpotifyClientCredentials(
        client_id=CLIENT_ID,
        client_secret=CLIENT_SECRET
    )
)

emociones_ia = {
    "feliz": ["feliz", "contento", "alegre", "motivado", "animado"],
    "triste": ["triste", "mal", "decaído", "sin ganas", "vacío"],
    "enojado": ["enojado", "molesto", "furioso", "ira", "rabia"],
    "relajado": ["relajado", "calmado", "tranquilo", "en paz"],
    "ansioso": ["ansioso", "estresado", "nervioso", "preocupado"]
}

def detectar_emocion(texto):
    t = texto.lower()
    t = re.sub(r"[^a-záéíóúñ ]", "", t)
    for emo, palabras in emociones_ia.items():
        if any(p in t for p in palabras):
            return emo
    return None

keywords = {
    "feliz": "música feliz latino pop positivo",
    "triste": "balada triste español melancólica",
    "enojado": "rock español fuerte enfadado",
    "relajado": "relax español acústico chill",
    "ansioso": "música calmante instrumental español"
}

def recomendar_musica(emocion):
    query = keywords[emocion]
    results = sp.search(q=query, type="track", limit=10)

    canciones = []
    for t in results["tracks"]["items"]:
        nombre = t["name"]
        artista = t["artists"][0]["name"]
        link = t["external_urls"]["spotify"]
        portada = t["album"]["images"][0]["url"]
        canciones.append([nombre, artista, portada, link])

    df = pd.DataFrame(canciones, columns=["Canción", "Artista", "Portada", "Link"])
    return HTML(df.to_html(
        escape=False,
        formatters={
            "Portada": lambda x: f'<img src="{x}" width="80">',
            "Link": lambda x: f'<a href="{x}" target="_blank">Escuchar</a>'
        }
    ))

while True:
    texto = input("¿Qué emoción sientes ahora?: ")
    emocion = detectar_emocion(texto)

    if emocion is None:
        print("\n No entendí tu emoción.")
        print("Intenta con: feliz, triste, enojado, relajado, ansioso\n")
    else:
        break

print(f"\n Detectando canciones para la emoción: {emocion.upper()}...\n")
display(recomendar_musica(emocion))
```
